```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
TNSchools = len(school_data_complete["School ID"].unique())
TNSchools
```

```python
#Calculate the total number of students
TNStudents = len(school_data_complete["Student ID"].unique())
TNStudents
```

```python
#Calculate the total budget
Total_Budget = school_data["budget"].sum()
Total_Budget
```

```python
#Calculate the average math score
Average_MS = student_data["math_score"].mean()
Average_MS
```

```python
#Calculate the average reading score
Average_RS = student_data["reading_score"].mean()
Average_RS
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_MS = school_data_complete.loc[school_data_complete["math_score"] >= 70]["math_score"].count()
passing_MS
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
passing_RS = school_data_complete.loc[school_data_complete["reading_score"] >= 70]["reading_score"].count()
passing_RS
```

```python
#Calculate the percentage of students who passed math for Overall Passing calculation
pass_math_percent = passing_MS/TNStudents*100
pass_math_percent
```

```python
#Calculate the percentage of students who passed reading for Overall Passing calculation
pass_reading_percent = passing_RS/TNStudents*100
pass_reading_percent
```

```python
#Overall Passing calculation
overall_passing = (Average_MS + Average_RS)/2
overall_passing
```

```python
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({
    "Total Schools": [TNSchools],
    "Total Students": [TNStudents],
    "Total Budget": [Total_Budget],
    "Average Math Score": [Average_MS],
    "Average Reading Score": [Average_RS],
    "% Passing Math":[pass_math_percent],
    "% Passing Reading": [pass_reading_percent],
    "% Overall Passing Rate": [overall_passing]})

#Format dataframe above for cleaner display view 
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.0f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2f}%".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map("{:,.2f}%".format)

#Print
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name 
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School name and type summary table by indexing School name 
schoolNS = school_data_complete.set_index("school_name").groupby(["school_name"])
school_types = school_data.set_index("school_name")["type"]
school_types
```

```python
#Student per school count by indexing school name
students_per_school = schoolNS['Student ID'].count()
students_per_school
```

```python
#Total School Budget sum by indexing school name
school_budget = school_data.set_index("school_name")["budget"]
school_budget
```

```python
#Per Student Budget count by indexing school name
per_student_budget = school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"]
per_student_budget
```

```python
#Calculate average Math and Reading scores 
average_math_score = schoolNS["math_score"].mean()

average_read_score = schoolNS["reading_score"].mean()
```

```python
#Calculate % Passing Math
passing_math_percent = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()/students_per_school*100 

#Calculate % Passing Reading
passing_read_percent = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/students_per_school*100

#Calculate % Overall Passing
overall_pass_rate = (passing_math_percent + passing_read_percent)/2

#Create a dataframe to hold the above results
summary_by_school = pd.DataFrame({
   "School Type": school_types,
   "Total Students": students_per_school,
   "Total School Budget": school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_read_score,
    "% Passing Math": passing_math_percent,
    "% Passing Reading": passing_read_percent,
    "% Overall Passing Rate": overall_pass_rate})

#Format dataframe to hold a cleaner display of the above results
summary_by_school["Total Students"] = summary_by_school["Total Students"].map("{:,}".format)
summary_by_school["Total School Budget"] = summary_by_school["Total School Budget"].map("${:,.0f}".format)
summary_by_school["Per Student Budget"] = summary_by_school["Per Student Budget"].map("${:,.0f}".format)
summary_by_school["Average Math Score"] = summary_by_school["Average Math Score"].map("{:,.2f}".format)
summary_by_school["Average Reading Score"] = summary_by_school["Average Reading Score"].map("{:,.2f}".format)
summary_by_school["% Passing Math"] = summary_by_school["% Passing Math"].map("{:,.2f}%".format)
summary_by_school["% Passing Reading"] = summary_by_school["% Passing Reading"].map("{:,.2f}%".format)
summary_by_school["% Overall Passing Rate"] = summary_by_school["% Overall Passing Rate"].map("{:,.2f}%".format)

#print
summary_by_school
```

```python
#Sort and display the top five performing schools by % overall passing
top_5 = summary_by_school.sort_values("% Overall Passing Rate", ascending = False)
top_5.head()
```

```python
#Sort and display the five worst-performing schools by % overall passing
bottom_5 = summary_by_school.sort_values("% Overall Passing Rate")
bottom_5.head()
```

```python
#Create bins
spending_ranges = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_data_complete["spending_ranges"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spending_ranges, labels = group_names)

#group by spending ranges
Spending = school_data_complete.groupby("spending_ranges")

#average math score
Math_Average = Spending["math_score"].mean()
#average reading score
Reading_Average = Spending["reading_score"].mean()
#% passing math
Passing_Math = school_data_complete[school_data_complete["math_score"] >=70].groupby("spending_ranges")["Student ID"].count()/Spending["Student ID"].count()*100
#% passing reading
Passing_Reading = school_data_complete[school_data_complete["reading_score"] >=70].groupby("spending_ranges")["Student ID"].count()/Spending["Student ID"].count()*100
#overall passing rate(average of above two)
Overall_Passing = (Passing_Math + Passing_Reading)/2

#Combine the series into a dataframe
scores_by_school_spend = pd.DataFrame({
    "Average Math Score": Math_Average,
    "Average Reading Score": Reading_Average,
    "% Passing Math": Passing_Math,
    "% Passing Reading": Passing_Reading,
    "% Overall Passing Rate": Overall_Passing})

#Format for cleaner display
scores_by_school_spend["Average Math Score"] = scores_by_school_spend["Average Math Score"].map("{:,.2f}".format)
scores_by_school_spend["Average Reading Score"] = scores_by_school_spend["Average Reading Score"].map("{:,.2f}".format)
scores_by_school_spend["% Passing Math"] = scores_by_school_spend["% Passing Math"].map("{:,.2f}%".format)
scores_by_school_spend["% Passing Reading"] = scores_by_school_spend["% Passing Reading"].map("{:,.2f}%".format)
scores_by_school_spend["% Overall Passing Rate"] = scores_by_school_spend["% Overall Passing Rate"].map("{:,.2f}%".format)


scores_by_school_spend
```
